```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
```

```python
#brand
brand=[]
for i in soup.find_all("div",class_="_4rR01T"):
    print(i.text)
    regex=re.findall("^[a-zA-Z]+",i.text)
    print(regex)
    if regex:
        brand.append(regex[0])
    else:
        brand.append(np.nan)
```

```python
len(brand)
```

```python
#display
display_size=[]
for i in soup.find_all("li",class_="rgWa7D"):
        text=i.text
        regex=re.findall("(\d+\.\d+) inch",text)
        if regex:
            display_size.append(regex[0])
        else:
            display_size.append(np.nan)
        
```

```python
print(display_size)
len(display_size)
```

```python
#discount
discount=[]
for i in soup.find_all("div",class_="_2kHMtA"):
    a=i.find("div",class_="_3Ay6Sb")      
    if a:
        regex = re.findall('^\d+',a.text)
        discount.append(regex[0])
    else:
        discount.append(np.nan)
```
